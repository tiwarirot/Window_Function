```python
df = pd.read_excel('Category_Increase.xlsx')
```

```python
df.head()
```

```python
# Step 1: Use shift function to move data 1 row forward
df['Lag_Value'] = df.groupby(['Category'])['Revenue'].shift(1).fillna(0)
df.head()
```

```python
# Step 2: Calculate difference between Revenue and Lag Value
df['Revenue_Diff'] = (df['Revenue']-df['Lag_Value'])
df.head()
```

```python
df_org = df.copy()
df_org
```

```python
# Step 3: Count total records under each category before Lag Function
df_org['Total_Year'] = df_org.groupby(['Category'])['Fiscal_Year'].transform('count')
df_org
```

```python
# Step 4: Count total records under each category after Lag Function
df_org['Incr_Rev'] = df_org[df_org.Revenue_Diff > 0].groupby(['Category'])['Fiscal_Year'].transform('count')
```

```python
# Step 5: Compare the count of records before and after lag function
df_org['Flag'] = (df_org['Total_Year']==df_org['Incr_Rev'])
df_org = df_org[df_org['Flag'] == True]
df_org
```

```python
# Step 6: Print the result
df_new = df_org['Category'].unique()
df_new
```
